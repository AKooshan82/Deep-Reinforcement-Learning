# Policies (5 points)

First define a greedy policy that takes a state $s$ and returns the action $a$ with the highest $Q(s,a)$.

```python
def greedy_policy(state: int, q_values: np.ndarray) -> int:
    # TODO: Select a greedy action
    action = np.argmax(q_values[state])
    return action
```

Now define a policy that selects a greedy action with probability $1-\varepsilon$ and selects a random action with probability $\varepsilon$, for given a constant $\varepsilon$.

```python
def epsilon_greedy_policy(state: int, q_values: np.ndarray, epsilon: float) -> int:
    # TODO: Select an epsilon-greedy action
    if np.random.rand() < epsilon:
        action = np.random.choice(q_values.shape[1])
        # action = random_policy(state, q_values)
    else:
        action = greedy_policy(state, q_values)

    return action
```

# Dyna-Q (25 points)

**Dyna-Q** combines model-free and model-based reinforcement learning by integrating direct learning from experience with planning from a learned model.
Unlike **Q-learning**, which only updates the action values based on real experiences, **Dyna-Q** uses a learned model of the environment to simulate additional experiences and update the action values, accelerating the learning process.

The environment we are working with is *deterministic* when `is_slippery=False`.
For this reason we use also use a *deterministic* model for the environment.
In order to see how **Dyna-Q** gets implemented for *stochastic* environmnets, check out the workshop!

## Planning (10 points)

In addition to this update from real experiences, **Dyna-Q** updates the action values using simulated experiences:


$$Q(s, a) \leftarrow Q(s, a) + \alpha \left(\hat{r} + \gamma \max_{a'} Q(\hat{s}, a') - Q(s, a)\right)$$


Complete the `q_planning` function to perform `n` steps of planning.

```python
def q_planning(model: dict, q: np.ndarray, alpha: float, gamma: float, n: int) -> np.ndarray:
    for _ in range(n):
        # TODO: Randomly sample a known state-action pair
        state, action = random.choice(list(model.keys()))
        next_state, reward = model[state, action]
        # TODO: Update Q-value using the deterministic transition
        q[state, action] += alpha * (reward + gamma * np.max(q[next_state]) - q[state, action])
    return q
```

## Learning (15 points)

For `n_episodes`, take ɛ-greedy actions based on the Q-values and at each step update them:

$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha \left(r_t + \gamma \max_{a} Q(s_{t+1}, a) - Q(s_t, a_t)\right)$$

After each step you must also update the *deterministic* model and call `q_planning` to perform the planning steps.

```python
def dyna_q(n_episodes: int, env: gym.Env, epsilon: float, alpha: float,
           gamma: float, n: int) -> tuple[np.ndarray, np.ndarray]:
    """Dyna-Q algorithm for deterministic environments."""
    reward_sums = np.zeros(n_episodes)
    q = np.zeros((env.observation_space.n, env.action_space.n))
    # Dyna-Q model for deterministic environments
    model = defaultdict(dict)
    for episode_i in (pbar := trange(n_episodes, leave=False)):
        state, info = env.reset()
        reward_sum, terminal = 0, False
        while not terminal:
            # TODO: Take ɛ-greedy action
            action = epsilon_greedy_policy(state, q, epsilon)
            # Take step in environment
            next_state, reward, terminal, truncated, info = env.step(action)
            # TODO: Q-learning update
            best_next_q = np.max(q[next_state])
            q[state, action] += alpha * (reward + gamma * best_next_q - q[state, action])
            # TODO: Update deterministic model
            model[(state, action)] = (next_state, reward)
            # TODO: Planning step(s)
            q = q_planning(model, q, alpha, gamma, n)
            # TODO: Move to next state
            state = next_state
            # TODO: Update reward sum
            reward_sum += reward
        pbar.set_description(f'Episode Reward {int(reward_sum)}')
        reward_sums[episode_i] = reward_sum
    return q, reward_sums
```

### 1. Introduce a Baseline for Q-values

* **Method:** Instead of initializing all Q(s, a) values to zero or small random numbers, initialize them to a more optimistic or informed baseline. For example, initializing to a value slightly higher than the maximum possible reward, or using a simple heuristic based on the state (e.g., distance to goal).
* **Issue Addressed:** Slow initial exploration and convergence, especially in environments with sparse or delayed rewards. Zero-initialization can make the agent overly pessimistic initially, requiring many interactions before positive rewards propagate back significantly.
* **Justification:** Optimistic initialization encourages the agent to explore unknown state-action pairs more readily. By assuming unknown actions might lead to better outcomes, the agent is incentivized to try them at least once. ### 2. Change the value of ε as a function of episode progression (ε-decay or annealing)

* **Method:** Start with a high value of ε (e.g., 1.0 or 0.9) and gradually decrease it over episodes or timesteps towards a small floor value (e.g., 0.1 or 0.01). Common decay schedules include linear decay, exponential decay, or step-wise decay.
* **Issue Addressed:** The static exploration-exploitation trade-off. A fixed high ε leads to perpetual random actions, preventing the agent from fully exploiting its learned knowledge. A fixed low ε might lead to premature convergence to a suboptimal policy because the agent didn't explore enough initially.
* **Justification:** This dynamic approach balances exploration and exploitation over time. High exploration (high ε) at the beginning is crucial for discovering the environment and potentially good policies. As the agent learns more and its Q-values become more reliable, reducing ε allows it to increasingly exploit this knowledge, leading to more stable and optimal behavior later in training. This avoids both insufficient exploration early on and excessive randomness later.

### 3. Replace the ε-greedy policy with a different policy

* **Method:** Use alternative action-selection strategies like **Softmax (Boltzmann)** exploration or **Upper Confidence Bound (UCB)** policies.
* **Issue Addressed:** The inefficiency of **ε-greedy's** exploration. When **ε-greedy** explores, it chooses a *random* action with equal probability among all non-greedy actions, regardless of their estimated values. This can waste time exploring clearly suboptimal actions.
* **Justification:**
    * **Softmax/Boltzmann:** Selects actions probabilistically based on their current Q-value estimates. Actions with higher Q-values are more likely to be selected, but actions with lower Q-values still have a chance (controlled by a temperature parameter). This focuses exploration on more promising alternatives rather than purely random ones.
    * **UCB:** Explicitly balances exploration and exploitation by selecting actions that have a high estimated value *or* have high uncertainty (i.e., haven't been tried often). This provides a more principled way to explore actions that could potentially be optimal but haven't been sampled enough.
    Both alternatives can lead to more directed and potentially faster learning by exploring more intelligently than **ε-greedy**.

### 4. Change the number of planning steps as a function of episode progression (Adaptive Planning / Dyna-Q variants)

* **Method:** Applicable in model-based RL approaches like **Dyna-Q**. Instead of performing a fixed number (n) of planning steps (learning from the model) after each real interaction, adjust 'n' dynamically. For example, increase 'n' when the model seems accurate or when a surprising event occurs (high TD error), or decrease 'n' early when the model is poor.
* **Issue Addressed:** Inefficiency of fixed planning effort. Early in training, the learned model of the environment might be inaccurate, making extensive planning potentially misleading or wasteful. Later, a fixed small 'n' might not be enough to quickly propagate important updates through the value function when the model *is* good or when the environment changes subtly. Computational cost is also a factor.
* **Justification:** Adapting the planning steps allows for more efficient use of computation. More planning can be done when it's most beneficial (e.g., after experiencing a state with a large value change, suggesting that related states also need updating) or when the model is reliable. Less planning can be done when the model is likely inaccurate or when the agent is operating in well-known parts of the state space. This optimizes the trade-off between learning speed and computational cost.

```python
def boltzmann_action_selection(state, q_values, temperature=0.1):
    """Selects an action using the Boltzmann (Softmax) distribution.
    Args:
        state: The current state.
        q_values: A NumPy array representing the Q-value function.
        temperature: Controls the randomness of action selection. Lower values
            make the policy more greedy, higher values make it more exploratory.
            Defaults to 0.1.

    Returns:
        The selected action.
    """
    preferences = q_values[state] / temperature if temperature > 0 else q_values[state]
    if np.all(preferences == preferences[0]):  # Handle cases where all preferences are equal
        return np.random.randint(len(q_values[state]))
    exp_preferences = np.exp(preferences - np.max(preferences))
    probabilities = exp_preferences / np.sum(exp_preferences)
    return np.random.choice(len(q_values[state]), p=probabilities)
def adaptive_planning_horizon(episode, initial_steps=5, growth_rate=1.02, max_steps=200):
    """Determines the number of planning steps based on the current episode.
    Args:
        episode: The current episode number (0-indexed).
        initial_steps: The minimum number of planning steps.
        growth_rate: The factor by which the planning steps can increase.
        max_steps: The maximum number of planning steps allowed.

    Returns:
        The number of planning steps for the current episode.
    """
    steps = int(initial_steps * (growth_rate ** episode))
    return min(steps, max_steps)
def epsilon_schedule(epsilon_0, episode, decay_factor=0.01):
    return epsilon_0 / (1 + decay_factor * episode)
def dyna_q_improved(n_episodes: int, env: gym.Env, epsilon_0: float, alpha: float,
                    gamma: float, n: int) -> tuple[np.ndarray, np.ndarray]:
    reward_sums = np.zeros(n_episodes)
    q = np.full((env.observation_space.n, env.action_space.n), 0.5)

    model = defaultdict(dict)

    for episode_i in (pbar := trange(n_episodes, leave=False)):
        state, info = env.reset()
        reward_sum, terminal = 0, False

        epsilon = epsilon_schedule(epsilon_0, episode_i)

        while not terminal:
            action = boltzmann_action_selection(state, q, temperature=0.1)

            next_state, reward, terminal, _, _ = env.step(action)
            best_next_action = np.argmax(q[next_state])
            q[state, action] += alpha * (reward + gamma * q[next_state, best_next_action] - q[state, action])

            model[state, action] = (reward, next_state)

            planning_steps = adaptive_planning_horizon(episode_i, initial_steps=5, growth_rate=1.01, max_steps=200)
            q = q_planning(model, q, alpha, gamma, int(planning_steps))

            state = next_state

            reward_sum += reward

        pbar.set_description(f'Episode Reward {int(reward_sum)}')
        reward_sums[episode_i] = reward_sum

    return q, reward_sums
# set for reproducibility
np.random.seed(2025)

# parameters needed by our policy and learning rule
params = {
    'epsilon_0': 1.0,     # epsilon-greedy policy (10% exploration)
    'alpha': 0.05,       # learning rate (0.1 is a common value)
    'gamma': 0.99,       # temporal discount factor (common value, indicating future rewards are important)
    'n': 50,             # number of planning steps (can vary, 50 is a reasonable starting point)
}

# episodes/trials
n_episodes = 5000

# environment initialization
env = gym.make("FrozenLake-v1", map_name="8x8", is_slippery=False)

# Solve Frozen Lake using Dyna-Q
value_dyna_q, reward_sums_dyna_q = dyna_q_improved(n_episodes, env, **params)

# Plot the results
# Assuming plot_performance function is defined elsewhere and works correctly
plot_performance(env, value_dyna_q, reward_sums_dyna_q)
```

# Reward Shaping (10 points)

As you know, [Reward Function Design is Difficult](https://www.alexirpan.com/2018/02/14/rl-hard.html#reward-function-design-is-difficult).
However there a few main principles than can help speed up *Temporal-Difference Learning*.
Here we ask you to modify the reward function to achieve better results.

```python
from gymnasium.envs.toy_text import FrozenLakeEnv

class CustomFrozenLakeEnv(FrozenLakeEnv):
    def step(self, action):
        obs, reward, terminated, truncated, info = super().step(action)
        # Modify the reward calculation
        reward = self.custom_reward_function(obs, reward, terminated)
        return obs, reward, terminated, truncated, info

    def custom_reward_function(self, observation, reward, done):
        # TODO: Define your custom reward logic here
        goal_state = self.observation_space.n - 1
        distance_to_goal = abs(observation % 8 - goal_state % 8) + abs(observation // 8 - goal_state // 8)
        if observation == goal_state:
            custom_reward = 10.0
        elif done:
            custom_reward = -5.0
        else:
            custom_reward = (1.0 / (distance_to_goal + 1) - 1) * 0.01
        return custom_reward
```

`Your Answer:`
* **Reward Shaping:**
    * **Method:** Augmenting the sparse environmental reward with additional, more frequent reward signals designed to guide the agent. **Potential-based reward shaping** is a common technique to ensure the optimal policy remains unchanged.
    * **Issue Addressed:** Very slow learning in environments with sparse rewards (e.g., reward only upon reaching a distant goal). The agent may wander aimlessly for a long time without receiving any positive feedback.

# Prioritized Sweeping (20 points)

In the Dyna agents presented in the preceding sections, simulated transitions are started in state-action pairs selected uniformly at random from all previously experienced pairs.
But a uniform selection is usually not the best; planning can be much more efficient if simulated transitions and updates are focused on particular state-action pairs.

## Priority Planning (10 points)

The `q_planning_priority` function performs planning updates using **Prioritized Sweeping** by first checking if the priority queue is empty and then iterating for a specified number of planning steps.
It pops the state-action pair with the highest priority from the queue, retrieves the deterministic transition, and updates the Q-value using the temporal-difference (TD) error.
Next, it updates priorities for predecessor state-action pairs that transition to the current state, calculating their TD errors and pushing significant ones back into the priority queue.

```python
def q_planning_priority(model: dict, q: np.ndarray, priorities: list, alpha: float, gamma: float, n: int) -> np.ndarray:
    """Performs planning updates using Prioritized Sweeping."""

    for _ in range(n):
        if not priorities:
            break  # No priorities to process

        # Sample the state-action pair with the highest priority
        priorities.sort(key=lambda x: x[0], reverse=True)
        priority, state, action = priorities.pop(0)

        # Retrieve deterministic transition
        if (state, action) not in model:
            continue  # Skip if the state-action pair is not in the model
        reward, next_state = model[(state, action)]

        # Update Q-value using the deterministic transition
        old_q = q[state, action]
        q[state, action] += alpha * (reward + gamma * np.max(q[next_state]) - q[state, action])
        delta_q = abs(q[state, action] - old_q)

        # Update priorities for predecessors
        for prev_state, prev_action in model:
            if model[(prev_state, prev_action)][1] == state:  # If this (prev_state, prev_action) leads to the current state
                prev_reward = model[(prev_state, prev_action)][0]
                priority_update = abs(prev_reward + gamma * np.max(q[state]) - q[prev_state, prev_action])
                # Add or update priority in the list
                found = False
                for i, (p, s, a) in enumerate(priorities):
                    if s == prev_state and a == prev_action:
                        priorities[i] = (max(p, priority_update), s, a)
                        found = True
                        break
                if not found and priority_update > 0:
                    priorities.append((priority_update, prev_state, prev_action))

    return q
```

## Learning with Prioritized Sweeping (10 points)

Similar to the original `dyna_q` function, you need to implement the `dyna_q_priority`.
This time you must also store priorities and use a threshold parameter $\theta$ that determines whether a state-action pair should be added to the priority queue based on the magnitude of the temporal-difference (TD) error.

Specifically, $\theta$ is used to filter out small TD errors that are considered insignificant.
When the absolute value of the TD error for a state-action pair is greater than $\theta$, the pair is added to the priority queue.
This ensures that only state-action pairs with significant TD errors are prioritized for planning updates.

```python

def dyna_q_priority(n_episodes: int, env: gym.Env, epsilon: float, alpha: float,
                    gamma: float, n: int, theta: float) -> tuple[np.ndarray, np.ndarray]:
    """Dyna-Q with Prioritized Sweeping algorithm for deterministic environments."""

    reward_sums = np.zeros(n_episodes)
    q = np.zeros((env.observation_space.n, env.action_space.n))

    # Dyna-Q model for deterministic environments
    model = defaultdict(dict)

    # Priority queue for prioritized sweeping
    priorities = []

    for episode_i in (pbar := trange(n_episodes, leave=False)):
        state, info = env.reset()
        reward_sum, terminal = 0, False

        while not terminal:
            # Take ɛ-greedy action
            action = epsilon_greedy_policy(state, q, epsilon)

            # Take action and observe reward and next state
            next_state, reward, terminal, _, _ = env.step(action)

            # Q-learning update
            old_q_sa = q[state, action]
            q[state, action] += alpha * (reward + gamma * np.max(q[next_state]) - q[state, action])
            td_error = abs(q[state, action] - old_q_sa)

            # Update deterministic model
            model[state, action] = (reward, next_state)

            # Update priority queue if the TD error is significant
            if td_error > theta:
                priorities.append((td_error, state, action))

            # Planning step with prioritized sweeping
            q = q_planning_priority(model, q, priorities, alpha, gamma, n, theta)

            # Move to next state
            state = next_state

            # Update reward sum
            reward_sum += reward

        pbar.set_description(f'Episode Reward {int(reward_sum)}')
        reward_sums[episode_i] = reward_sum

    return q, reward_sums
```
